```python
import pandas as pd
```

```python
df=pd.read_csv('https://press.radiocut.fm/bicycle-travels-jan-2018.csv',sep=',')
```

```python
df['pickup_time']=df.ix[:,'pickup_time'].apply(lambda x : x.split(" ")[1].split(":")[0])
```

```python
df
```

```python
df=df[['source_station_name','pickup_time','source_station_code']]
df['pickup_time']=pd.to_numeric(df['pickup_time'])
```

```python

lower_filter=(df['pickup_time']>5)
upper_filter=(df['pickup_time']<12)
df=df[lower_filter]
df=df[upper_filter]
df['item']=1
df
```

```python
df.groupby(by='source_station_name')['item'].count().sort_values(ascending=False).head(5)
```
